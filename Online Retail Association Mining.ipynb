Association Mining of items for country France

```python
transactions_france=data[['InvoiceNo','Description','Quantity']].loc[data['Country']=='France']
```

```python
transactions_france=transactions_france.loc[transactions_france['Quantity']>0]
```

```python
transactions_france.groupby(['InvoiceNo','Description']).agg({"Quantity": "sum"}).head()
```

```python
transactions_france['Quantity']=1
```

```python
transactions_france.head()
```

```python
table=pd.pivot_table(transactions_france,values='Quantity',index=['InvoiceNo'],columns=['Description'])
```

```python
table
```

```python
t=[]
for i in range(0,392):
   t.append([table.columns[j] for j in range(0,1563) if table.iloc[i][j]==1]) 
```

```python
t
```

Applying Apriori algorithm

```python
from apyori import apriori
rules=apriori(t, min_support=0.003, min_confidence=0.2, min_lift=3, max_length=3)
```

```python
results=list(rules)
```

```python
results
```

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1]) for result in results]
    supports = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences,lifts))

resultsinDataFrame= pd.DataFrame(inspect(results), columns=['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
resultsinDataFrame
```
