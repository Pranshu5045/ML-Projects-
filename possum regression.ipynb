```python
x=data.drop('age',axis=1)
y=data['age']
```

```python
from feature_engine.selection import DropConstantFeatures,DropDuplicateFeatures,SmartCorrelatedSelection
```

```python
from sklearn.pipeline import Pipeline
pipe=Pipeline([('Constant',DropConstantFeatures(tol=0.9)),('Duplicate',DropDuplicateFeatures()),('Correlated',SmartCorrelatedSelection(estimator=SVR()))])
```

```python
X=pipe.fit_transform(x)
```

```python
X.shape
```

```python
x.shape
```

Only 'site' feature was dropped

removing outliers

```python
Q1 = X.quantile(0.25)
Q3 = X.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
X = X[~((X < (Q1 - 1.5 * IQR)) |(X > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
X
```

```python
y=y[X.index]
```

```python
from sklearn.preprocessing import LabelEncoder
en=LabelEncoder()
for i in X.columns:
    if X[i].dtype=='object':
        X[i]=en.fit_transform(X[i])
```

```python
Y=en.fit_transform(y)
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
sca=StandardScaler()
X_new=sca.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X_new,Y)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain,ytrain)
lrpred=lr.predict(xtest)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print('Root mean squared error: ',np.sqrt(mean_squared_error(ytest,lrpred)))
print('Mean absolute error: ',mean_absolute_error(lrpred,ytest))
print('R2 score: ',r2_score(lrpred,ytest))
```

```python
sns.residplot(x=ytest,y=lrpred);
```

```python
from sklearn.model_selection import cross_validate
resultsl=cross_validate(estimator=LinearRegression(),cv=10,X=X_new,y=Y,scoring=['neg_mean_squared_error','r2'])
```

```python
print('mean_squared_error: ',np.mean(-resultsl['test_neg_mean_squared_error']))
print('r2 score: ',resultsl['test_r2'].mean())
```

## SVR

```python
from sklearn.model_selection import GridSearchCV
param = {'kernel' : ('linear', 'poly', 'rbf', 'sigmoid'),'C' : [1,5,10],'degree' : [3,8],'coef0' : [0.01,10,0.5],'gamma' : ('auto','scale')}
grid=GridSearchCV(estimator=SVR(),param_grid=param,cv=10,verbose=1,scoring='neg_mean_squared_error')
grid.fit(xtrain,ytrain)
print(grid.best_estimator_)
```

```python
sv=SVR(C=1, coef0=0.01, gamma='auto')
sv.fit(xtrain,ytrain)
predsv=sv.predict(xtest)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print('Root mean squared error: ',np.sqrt(mean_squared_error(ytest,predsv)))
print('Mean absolute error: ',mean_absolute_error(ytest,predsv))
print('R2 score: ',r2_score(ytest,predsv))
```

```python
sns.residplot(x=ytest,y=predsv);
```

```python
from sklearn.model_selection import cross_validate
resultsv=cross_validate(estimator=SVR(),cv=10,X=X_new,y=Y,scoring=['neg_mean_squared_error','r2'])
```

```python
print('mean_squared_error: ',np.mean(-resultsv['test_neg_mean_squared_error']))
print('r2 score: ',resultsv['test_r2'].mean())
```
